## Estutura da Base de dados

```python
def diagnostico(df, nome="DataFrame"):
 
    print("=" * 55)
   
    print(f"   DIAGNÓSTICO: {nome}")
    print("=" * 55)
    print(f"  Linhas:           {df.shape[0]:,}")
    print(f"  Colunas:          {df.shape[1]}")
    print(f"  Linhas duplicadas:{df.duplicated().sum():,}")
    print()
    
    nulos = df.isnull().sum()
    pct   = (nulos / len(df) * 100).round(1)
    
    print("  Coluna           | Tipo       | Nulos | % Nulos")
    print("  " + "-"*50)
    for col in df.columns:
        tipo = str(df[col].dtype)
        print(f"  {col:<18}| {tipo:<10} | {nulos[col]:<5} | {pct[col]}%")
    print("=" * 55)


diagnostico(df, "Base Varejo")
```

### Analise das duplicatas

```python
print(f"Linhas completamente iguais: {df.duplicated().sum()}")

print(f"Compras duplicadas (CO_ID): {df.duplicated(subset=['CO_ID']).sum()}")

print(df[df.duplicated(subset=['CO_ID'], keep=False)].head())
```

### Validação da Regra de Negócio do Identificador de Compra (CO_ID)

```python
print(f"Quantidade de registros: {len(df)}")
print(f"Quantidade de compras únicas: {df['CO_ID'].nunique()}")
print(f"Registros com CO_ID repetido: {df.duplicated(subset=['CO_ID']).sum()}")
```

A base possui 733.447 registros e 18.471 compras únicas identificadas pela coluna CO_ID.

Foram encontrados 714.976 registros com CO_ID repetido. Entretanto, essa repetição não caracteriza necessariamente uma duplicata, pois uma mesma compra pode conter vários produtos diferentes registrados em linhas distintas.

Dessa forma, os registros com CO_ID repetido foram mantidos na base, sendo removidas apenas as linhas completamente idênticas identificadas durante a etapa de tratamento de duplicatas.

### Remoção de duplicatas

```python
# Verificando duplicatas reais (linhas totalmente iguais)
print(f"Linhas completamente iguais: {df.duplicated().sum()}")
print()

# Visualizando algumas duplicatas (caso existam)
print("Exemplo de linhas duplicadas:")
print(df[df.duplicated(keep=False)].head())
print()

# Removendo duplicatas
df = df.drop_duplicates()

print("Após remoção:")
print(f"Linhas duplicadas restantes: {df.duplicated().sum()}")
print(f"Linhas finais do dataset: {df.shape[0]}")
```

Foram identificadas linhas completamente duplicadas na base de dados.  
Esses registros foram removidos para evitar distorções nas análises.

## Limpeza da categoria 

```python
if df['PR_CAT'].isnull().sum()> 0:
    df['PR_CAT'] = df['PR_CAT'].fillna("SEM CATEGORIA")
    print("Nulos em 'PR_CAT' preenchidos com 'SEM CATEGORIA'")

elif (df['PR_CAT'] == '#N/D').sum()>0:
    df['PR_CAT'] = df['PR_CAT'].replace('#N/D', 'SEM CATEGORIA')
    print("Valores '#N/D' em 'PR_CAT' substituídos por 'SEM CATEGORIA'")

else:
    print("PR_CAT já esta preenchida corretamente")
    
```

### Transformando data de string para datetime


```python
df['DATA'] = pd.to_datetime(df['DATA'], dayfirst=True, errors='coerce')
```
